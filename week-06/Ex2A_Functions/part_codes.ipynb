# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings – part codes

```python
supplier:productcode-size
```

```python
def get_supplierCode(variable):
    return variable.split(":")[0]
```

```python
def get_productNum(variable):
    return variable.split(":")[1].split("-")[0]
```

```python
def get_size(variable):
    return variable.split("-")[1]
```

```python
acme = "Acme:123-L"
```

```python
di = "DI:12345-M"
```

```python
ace = "ACE:1-12"
```

```python
print(get_supplierCode(acme))
```

```python
print(get_productNum(acme))
```

```python
print(get_size(acme))
```

```python
print(get_supplierCode(di))
```

```python
print(get_productNum(di))
```

```python
print(get_size(di))
```

```python
print(get_supplierCode(ace))
```

```python
print(get_productNum(ace))
```

```python
print(get_size(ace))
```

```python
parts = [acme, di, ace]
```

```python
count = 0
```

```python
for i in parts:
    count += 1
    print(f"{count}. The supplier code is: {get_supplierCode(i)}. The product number is {get_productNum(i)}. The size is {get_size(i)}.")
```
