Employee data This lab exercise makes updates to information extracted from an employee database.

```python
hr_list = [ ('0123', 'HR', 'Rebecca Yang', '69000'), 
            ('0121', 'IT', 'Mark Blick', '67000'), 
            ('0068', 'IT', 'Kahna Larsen', '112000'), 
            ('0234', 'OPS', 'Jim Smith', '54000') 
          ]
```

```python
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
```

```python
print(hr_list)
```

```python
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000') 
```

```python
print(hr_list)
```

```python
count = 0
```

```python
for i in hr_list:
    count += 1
    employeenum = i[0] 
    deptCode = i[1]
    name = i[2]
    income = i[3]
    print(f"{count}. {employeenum} | {deptCode} | {name} | {int(income):,}")
```
