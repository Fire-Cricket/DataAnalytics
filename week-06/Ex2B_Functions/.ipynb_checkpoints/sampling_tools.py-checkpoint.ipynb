```python
import random
```

```python
 products = ['Laptop', 
             'Monitor', 
             'Keyboard', 
             'Mouse', 
             'Webcam', 
             'Headset', 
             'Docking Station', 
             'USB Hub', 
             'Desk Lamp', 
             'Surge Protector'
            ]
```

```python
print(f" Random choice of the day: {random.choice(products)}")
```

```python
print(f" Usability survey: {random.sample(products, k=3)}")
```

```python
random.shuffle(products)
```

```python
print(f" Shuffled list: {products}")
```

```python
print(f"Todays transactions is: {random.randint(50, 300)}")
```
