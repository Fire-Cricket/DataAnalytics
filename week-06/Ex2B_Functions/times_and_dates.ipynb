```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
print(f"Today is {today.strftime('"%A, %B %d, %Y"')}")
```

# My Birthday 

```python
birthday = datetime.date(2005, 0o2, 24)
```

```python
print(f"My birthday is {birthday.strftime("%m/%d/%Y")}")
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f"90 days from today is {ninety_d.strftime('%m %d, %Y')}")
```

# Time for Dinner

```python
dinner_time = datetime.time(18, 30)
```

```python
print(f"Let’s meet for dinner at {dinner_time.strftime('%I:%M %p')}")
```
